```python
#import the pandas 
import pandas as pd
```

```python
#import the file using pandas 
df=pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
#information of data contain data type
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df["Review Rating"]=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#with good understand of coumname we convert to python snake
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#after convert to lower
df.columns
```

```python
#create a columns age_group
labels=['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#cerate the frequency of puchasind days 
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quaterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchases_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchases_frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```
